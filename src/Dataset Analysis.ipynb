```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# 60-20-20 Dataset

```python
train_df = pd.read_csv('/Users/nfanlo/dev/spanish-classifier-tfg/dataset/60-20-20/train.csv')
dev_df = pd.read_csv('/Users/nfanlo/dev/spanish-classifier-tfg/dataset/60-20-20/dev.csv')
test_df = pd.read_csv('/Users/nfanlo/dev/spanish-classifier-tfg/dataset/60-20-20/test.csv')
```

```python
pie_train=pd.DataFrame(train_df['sentiment'].replace(0,'Negativo').replace(1,'Neutro').replace(2,'Positivo').value_counts())
pie_train.reset_index(inplace=True)
pie_train.plot(kind='pie', title='Train Dataset (60%) de las Etiquetas Sentimientos',y = 'sentiment', 
          autopct='%1.1f%%', shadow=False, labels=pie_train['index'], legend = False, fontsize=14, figsize=(8,8))
```

```python
pie_dev=pd.DataFrame(dev_df['sentiment'].replace(0,'Negativo').replace(1,'Neutro').replace(2,'Positivo').value_counts())
pie_dev.reset_index(inplace=True)
pie_dev.plot(kind='pie', title='Dev Dataset (20%) de las Etiquetas Sentimientos',y = 'sentiment', 
          autopct='%1.1f%%', shadow=False, labels=pie_dev['index'], legend = False, fontsize=14, figsize=(8,8))
```

```python
pie_test=pd.DataFrame(test_df['sentiment'].replace(0,'Negativo').replace(1,'Neutro').replace(2,'Positivo').value_counts())
pie_test.reset_index(inplace=True)
pie_test.plot(kind='pie', title='Test Dataset (20%) de las Etiquetas Sentimientos',y = 'sentiment', 
          autopct='%1.1f%%', shadow=False, labels=pie_test['index'], legend = False, fontsize=14, figsize=(8,8))
```
